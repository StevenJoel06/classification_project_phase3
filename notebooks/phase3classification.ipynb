### A Critical Challenge: Class Imbalance
One of the most significant challenges identified immediately was the severe imbalance in our target variable.

As shown, approximately 84% of individuals do NOT have diabetes (Class 0), while only around 15% of individuals DO have diabetes (Class 1).

Why this matters: If a model simply predicts 'No Diabetes' for everyone, it would achieve 84% accuracy. However, it would completely fail to identify actual diabetes cases, which is unacceptable for our goal of early risk identification. Addressing this imbalance is a cornerstone of our project.

## 5. Data Preparation & Preprocessing

Step 1: Handling Duplicate Records
We identified and removed 24,206 duplicate rows, resulting in 229,474 unique records for our analysis.

Step 2: Separating Features and Target Variable
We define our independent variables (features, X) and our dependent variable (target, y).

Step 3: Splitting Data into Training and Testing Sets
To evaluate our model's performance on unseen data, we divide our dataset into two parts:

Training Set (80%): Used to teach the model patterns in the data.
Testing Set (20%): Used to evaluate how well our trained model performs on data it has never seen before, simulating real-world application.

```python
# --- Step 1: Handle Duplicate Rows ---
# Given the large dataset derived from survey responses, duplicates are possible.
initial_rows = len(df)
df.drop_duplicates(inplace=True)
rows_after_duplicates = len(df)
print(f"\nNumber of duplicate rows removed: {initial_rows - rows_after_duplicates}")
print(f"Number of rows after dropping duplicates: {rows_after_duplicates}")
```

```python
# --- Step 2: Separate Features (X) and Target (y) ---
# 'Diabetes_binary' is our target variable, all other columns are features.
X = df.drop('Diabetes_binary', axis=1) # Features
y = df['Diabetes_binary']             # Target variable

print("\n--- Features (X) Head ---")
print(X.head())
print("\n--- Target (y) Head ---")
print(y.head())
```

```python
# --- Step 3: Feature Scaling ---
# All features will be scaled. This is crucial for Logistic Regression.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\n--- Scaled Features (X_scaled) Head ---")
print(X_scaled.head())
print("\n--- Scaled Features Descriptive Statistics (to check mean near 0, std near 1) ---")
print(X_scaled.describe())
```

```python
# --- Step 4: Split Data into Training and Testing Sets ---
# We'll use 80% for training and 20% for testing.
# 'random_state' ensures reproducibility.
# 'stratify=y' is crucial for imbalanced datasets, which this one is.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\n--- Data Split Summary ---")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print("\n--- Class distribution in y_train ---")
print(y_train.value_counts(normalize=True))
print("\n--- Class distribution in y_test ---")
print(y_test.value_counts(normalize=True)) # Should be very similar to y_train
```

## 6. Model Selection & Iterative Evaluation (Results)
Our Goal: Balancing Accuracy and Identifying True Diabetes Cases
Our primary objective was not just overall accuracy, but critically, to build a model that effectively identifies individuals who do have diabetes (the minority class). Missing a diabetes case (a 'False Negative') can have significant health implications.

Key Performance Metrics Explained
For our stakeholders, it's vital to understand the metrics beyond simple accuracy, especially in imbalanced classification.

Accuracy: The overall proportion of correct predictions. While useful, it can be misleading in imbalanced datasets.

Confusion Matrix: A table that shows where our model made correct and incorrect predictions:

True Negative (TN): Correctly predicted No Diabetes.

False Positive (FP): Incorrectly predicted Diabetes (a 'false alarm').

False Negative (FN): Incorrectly predicted No Diabetes (a 'missed diagnosis' - critical).

True Positive (TP): Correctly predicted Diabetes.

Precision (for Diabetes): Out of all cases predicted as Diabetes, how many were actually Diabetes. High precision reduces false alarms.

Recall (for Diabetes - Critical for our goal): Out of all actual Diabetes cases, how many did our model correctly identify. High recall minimizes missed diagnoses.

F1-Score (for Diabetes): A balance between Precision and Recall.

AUC (Area Under the Receiver Operating Characteristic Curve): Measures the model's ability to distinguish between classes across all possible classification thresholds. A higher AUC (closer to 1) indicates better discriminative power.

### Model 1: Logistic Regression (Our Baseline)
We started with Logistic Regression, a simple yet powerful linear model, as our baseline due to its interpretability and efficiency. 

```python
# --- Choose and Train a Classification Model (Logistic Regression) ---
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42, solver='saga', max_iter=1000, class_weight='balanced') # Added class_weight='balanced'

print("\n--- Training the Logistic Regression model ---")
model.fit(X_train, y_train)
print("Model training complete!")
```

```python
# --- Make Predictions on the Test Set ---
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)

print("\n--- First 5 predictions (class labels) ---")
print(y_pred[:5])
print("\n--- First 5 prediction probabilities (for class 0 and class 1) ---")
print(y_pred_proba[:5])
```

```python
# --- Evaluate Model Performance ---
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Model Evaluation ---")

# A. Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# B. Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted No Diabetes (0)', 'Predicted Diabetes (1)'],
            yticklabels=['Actual No Diabetes (0)', 'Actual Diabetes (1)'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# C. Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# D. ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"\nArea Under the ROC Curve (AUC): {roc_auc:.4f}")
```

```python
# --- Hyperparameter Tuning (Grid Search) ---
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression 
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'] # Using L2 regularization for Logistic Regression
}

# Ensure class_weight is consistent with the initial model to preserve imbalance handling
grid_search = GridSearchCV(
    LogisticRegression(solver='saga', max_iter=1000, random_state=42, class_weight='balanced'),
    param_grid,
    cv=5, # Using 5-fold cross-validation
    scoring='f1', # Still using F1-score as the primary metric for tuning due to imbalance
    n_jobs=-1 # Use all available CPU cores for faster computation
)

print("\n--- Starting Hyperparameter Tuning (Grid Search) ---")
# Training this will take longer due to dataset size and cross-validation
grid_search.fit(X_train, y_train)

print(f"Best parameters found by Grid Search: {grid_search.best_params_}")
print(f"Best cross-validation F1-score: {grid_search.best_score_:.4f}")

best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
y_pred_proba_tuned = best_model.predict_proba(X_test)


print("\n--- Evaluation of Tuned Model on Test Set ---")
print(f"Accuracy (Tuned Model): {accuracy_score(y_test, y_pred_tuned):.4f}")
print("\nClassification Report (Tuned Model):")
print(classification_report(y_test, y_pred_tuned))

fpr_tuned, tpr_tuned, _ = roc_curve(y_test, y_pred_proba_tuned[:, 1])
roc_auc_tuned = auc(fpr_tuned, tpr_tuned)
print(f"Area Under the ROC Curve (AUC) for Tuned Model: {roc_auc_tuned:.4f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr_tuned, tpr_tuned, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_tuned:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Tuned Model')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

### Model 2: Logistic Regression with SMOTE (Addressing Imbalance)
Building on our baseline, we applied SMOTE to the training data only to create a truly balanced training environment for Logistic Regression.

```python
# Import SMOTE
from imblearn.over_sampling import SMOTE
from collections import Counter
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns



print("Original class distribution in training data:")
print(Counter(y_train))

# Initialize SMOTE
# random_state for reproducibility
# k_neighbors can be adjusted, default is 5
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("\nClass distribution after SMOTE on training data:")
print(Counter(y_train_resampled))

# Train a new Logistic Regression model on the SMOTE-resampled data
# We can remove class_weight='balanced' now since SMOTE balances the data directly,
# but it's often still useful to keep it, or test without it. Let's try without first
# to see the direct effect of SMOTE.
smote_model = LogisticRegression(random_state=42, solver='saga', max_iter=1000)

print("\n--- Training Logistic Regression model with SMOTE-resampled data ---")
smote_model.fit(X_train_resampled, y_train_resampled)
print("SMOTE Model training complete!")

# Make predictions on the original (non-resampled) test set
y_pred_smote = smote_model.predict(X_test)
y_pred_proba_smote = smote_model.predict_proba(X_test)

print("\n--- Model Evaluation (SMOTE Model) ---")

# Accuracy Score
accuracy_smote = accuracy_score(y_test, y_pred_smote)
print(f"Accuracy (SMOTE Model): {accuracy_smote:.4f}")

# Confusion Matrix
conf_matrix_smote = confusion_matrix(y_test, y_pred_smote)
print("\nConfusion Matrix (SMOTE Model):")
print(conf_matrix_smote)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_smote, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted No Diabetes (0)', 'Predicted Diabetes (1)'],
            yticklabels=['Actual No Diabetes (0)', 'Actual Diabetes (1)'])
plt.title('Confusion Matrix (SMOTE Model)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report
print("\nClassification Report (SMOTE Model):")
print(classification_report(y_test, y_pred_smote))

# ROC Curve and AUC
fpr_smote, tpr_smote, thresholds_smote = roc_curve(y_test, y_pred_proba_smote[:, 1])
roc_auc_smote = auc(fpr_smote, tpr_smote)

plt.figure(figsize=(8, 6))
plt.plot(fpr_smote, tpr_smote, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_smote:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve (SMOTE Model)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"\nArea Under the ROC Curve (AUC) (SMOTE Model): {roc_auc_smote:.4f}")
```

## 7. Best Model Analysis, Limitations, & Recommendations
Our Best Performing Model: Decision Tree Classifier with SMOTE
Based on our evaluation, the Decision Tree Classifier trained with SMOTE-resampled data stands out as our best performing model for this project, particularly when considering the crucial goal of identifying actual diabetes cases. While its overall accuracy of 78.87% is slightly lower than the Random Forest's (without SMOTE), its performance on the minority 'Diabetes' class is significantly better and more balanced, which is paramount in a medical screening context.

Key Performance Highlights (Decision Tree + SMOTE):

Recall for Diabetes (Class 1): 0.57
This means our model successfully identified 57% of all actual diabetes cases in the test set. For a medical screening tool, minimizing missed diagnoses (False Negatives) is critical, as a missed diabetes case can lead to severe health complications.

F1-Score for Diabetes (Class 1): 0.44
The F1-score, which balances precision and recall, indicates a reasonable trade-off. While the precision (0.36) means about one-third of our 'Diabetes' predictions were false alarms (False Positives), this is often an acceptable trade-off in early screening to ensure high recall of critical conditions.

AUC Score: 0.79
The Area Under the ROC Curve indicates good discriminative power, showing the model's ability to differentiate between individuals with and without diabetes.

Limitations of Our Model and Approach
It's important to acknowledge the limitations to set realistic expectations for stakeholders:

False Negatives Exist: Despite our efforts, the model still generates 3,052 False Negatives. This means a significant number of individuals with diabetes could still be missed. The implication is that this model is a screening tool, not a definitive diagnostic one, and further clinical testing would always be necessary for flagged individuals.

False Positives Impact: The 7,105 False Positives could lead to unnecessary follow-up tests, anxiety, and healthcare costs for individuals who are not diabetic. Stakeholders need to consider the cost-benefit of these 'false alarms' versus the benefit of identifying more true cases.

Recommendations For Stakeholders: How to Use This Model
This model should be deployed as an initial screening tool. It can identify individuals who exhibit a higher risk profile for diabetes based on their health indicators. These flagged individuals can then be prioritized for further, more definitive medical testing and consultation, optimizing healthcare resource allocation.
